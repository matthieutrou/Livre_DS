# Rareté des matchs nuls

La Premier League est le championnat d'Angleterre de football où s'affrontent du mois d'août au mois de mai suivant, vingt clubs professionnels.

Les données proviennent du site http://www.football-data.co.uk/englandm.php
Nous utiliserons les données de la saison 2013-2014 jusqu'à la saison 2018-2019

Nous garderons en variable la date, le nom des équipes à domicile et à l'extérieur et leur nombre de buts marqués.

Les objectifs sont de :
- Etudier les scores obtenus pour les matchs "Everton-Liverpool" ; le nombre de matchs nuls est-il exceptionnel?
- Determinier et représenter la loi de probabilité des scores obtenus en Premier League

## Etape 1 : Matchs nuls

#### Lecture et traitement du jeu de données

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/foot_1league_2013_2014.csv')
df2 = pd.read_csv('./data/foot_1league_2014_2015.csv')
df3 = pd.read_csv('./data/foot_1league_2015_2016.csv')
df4 = pd.read_csv('./data/foot_1league_2016_2017.csv')
df5 = pd.read_csv('./data/foot_1league_2017_2018.csv')
df6 = pd.read_csv('./data/foot_1league_2018_2019.csv')
```

```python
T = df1.append(df2, ignore_index=True)
T = T.append(df3, ignore_index=True)
T = T.append(df4, ignore_index=True)
T = T.append(df5, ignore_index=True)
T = T.append(df6, ignore_index=True)

T.head()
```

```python
col = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG']
T = T[col]

T.rename(columns={'HomeTeam' : "DOM", 'AwayTeam' : "EXT", 
                  'FTHG' : "BDOM", 'FTAG' : "BEXT"}, inplace=True)

T.head()
```

```python
len(T)
```

Nous avons donc 2281 matchs joués 

#### Fréquence de matchs nuls

```python
len(T.query('BDOM == 0 and BEXT == 0')) # On regarde le nombre de 0-0
```

```python
len(T.query('BDOM == 0 and BEXT == 0'))/len(T)
# On regarde la fréquence de 0-0
```

- Il y a environ 7% de 0-0

```python
T.query('BDOM == BEXT') # Voici tout les matchs nuls
```

```python
p = len(T.query('BDOM == BEXT'))/len(T)
p
```

- Il y a environ 23% de matchs nuls

## Etape 2 : Matchs joués entre Everton et Liverpool

```python
S = T.query('DOM in ["Everton", "Liverpool"] and EXT in ["Everton", "Liverpool"]')
S
```

```python
S.query('BDOM == BEXT') #Matchs nuls entre Everton et Liverpool
```

```python
len(S.query('BDOM == BEXT'))
```

- Il y a eu 7 matchs nuls entre Everton et Liverpool en 12 rencontres

- On a p la probabilité d'obtenir un match nul (0.23) en 12 matchs
- On peut utiliser une loi binomiale selon ses paramètres.

```python
from scipy.stats import binom

X = binom(12, p)
s = 0
for k in range(7,12):
    s += X.pmf(k)

s    
```

- On a à peine 1% de chance d'obtenir au moins 7 matchs nuls en 12 confrontations

```python
# On pouvait calculer ce pourcentage autrement.
# En effet, on a calculé P(X>=7) mais on pouvait calculer 1 - P(X<6)

# Et P(X<k) peut se calculer avec X.cdf(k)

1 - X.cdf(6)
```

## Etape 3 : Distribution des scores

#### Table des scores

```python
R = T.groupby(['BDOM', 'BEXT']).count()
R = R[['Date']]
R.reset_index(inplace=True)
R  # Ici on a tout les scores différents des matchs et le nombre de fois où ils sont arrivés
```

- On voit qu'il y a eu 171 '0-0' et 173 '0-1'

#### Fréquence des scores

```python
R['Freq'] = R['Date']/R['Date'].sum()
R.head()
```

```python
R.sort_values(by='Freq', ascending=False, inplace=True)
R.head(10)
```
